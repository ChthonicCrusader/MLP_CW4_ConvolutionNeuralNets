```python
def plot_graph(err_train,acc_train,err_valid,acc_valid,name):
    path='./results/'+name
    val_epoch=range(1,len(err_train)+1)
    err_epoch=range(1,len(err_valid)+1)
    plt.grid()
    plt.plot(err_epoch,err_train,'-b+',label='Train Error')
    plt.plot(val_epoch,err_valid,'-ro',label='Validation Error')
    plt.xlabel('Epochs')
    plt.ylabel('Error')
    plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3,
           ncol=2, mode="expand", borderaxespad=0.)
    if name:
        plt.savefig(path+'_error.png')
    plt.show()
    plt.close()
    plt.grid()
    plt.plot(err_epoch,acc_train,'-b+',label='Train Accuracy')
    plt.plot(val_epoch,acc_valid,'-ro',label='Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3,
           ncol=2, mode="expand", borderaxespad=0.)
    #if name:    
    #    plt.savefig(path+'_accuracy.png')
    plt.show()
    plt.close() 
```
